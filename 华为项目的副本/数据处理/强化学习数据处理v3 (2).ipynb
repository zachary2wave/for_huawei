```python
import numpy as np
import pandas as pd
import time, datetime
import pytz
import math
```

```python
Radio = pd.read_csv("./data/Radio_data.csv")
```

```python
radioApnameList = Radio.apname.unique().tolist()
```

```python
Radio1 = Radio[Radio.band == 2].sort_values(by=["apname","ts"] , ascending=True)
```

```python
startTime = []
for i in range(len(radioApnameList)):
     startTime.append(Radio1[Radio1.apname == radioApnameList[i]].iloc[2,1])
```

```python
#最后六位起始时间不对，推测缺失直接删除这六个ap
startTime
```

```python
radioApnameList = radioApnameList[:-6]
```

```python
radioApnameList
```

```python
Radio2 = Radio1[Radio1.apname.isin(radioApnameList)]
```

```python
#单个ap从起始时间到结束时间应该有540个数据，多了或少了数据的ap直接删除
for i in range(len(radioApnameList)):
     print(radioApnameList[i],len(Radio2[Radio2.apname == radioApnameList[i]]))
```

```python
Radio3 = Radio2.sort_values(by=["ts","apname"] , ascending=True)
```

```python
abandonList = []
for i in range(len(radioApnameList)):
     if(len(Radio2[Radio2.apname == radioApnameList[i]])!=540):
        abandonList.append(radioApnameList[i])
```

```python
#需要删除的ap
abandonList
```

```python
for i in abandonList:
    radioApnameList.remove(i)
```

```python
#只选取功率，信道，信道利用率，同频干扰率交给侯坤林添加
Radio3 = Radio3.sort_values(by=["ts","apname"] , ascending=True)
Radio4 = Radio3[Radio3.apname.isin(radioApnameList)][["ts","apname","power","channel","channelusage","interferenceratio"]]
Radio4 = Radio4.sort_values(by=["apname"] , ascending=True)
Radio4 = Radio4[["apname","power","channel","channelusage","interferenceratio"]]
```

```python
ap = Radio4.iloc[[i*540 for i in range(151)]].apname.unique()
```

```python
ap = ap.reshape(151,1)
```

```python
ap.shape
```

```python
Radio4 = Radio4[["power","channel","channelusage","interferenceratio"]]
dataClear = []
for j in range(540):
   dataClear.append(np.array(Radio4.iloc[[i*540+j for i in range(151)]]))
```

```python
#最后以数组形式保存数据
data = np.array(dataClear)
```

```python
data
```

```python
data_mean = data.mean(axis=0)
data_mean
```

```python
data_mean.shape
```

```python
data_mean = np.hstack((ap,data_mean))
```

```python
data_mean
```

```python
colum = ["apname","power","channel","channelusage","interferenceratio"]
```

```python
df = pd.DataFrame(data_mean,columns = colum)
```

```python
df
```

```python
chanList = df.channel.unique().tolist()
```

```python
all = []
```

```python
for chan in chanList:
    lis = []
    for i in range (151) :
        if (df.iloc[i,2]==chan):
            lis.append(np.array(df.iloc[i,:]))
    all.append(lis)
            
        
    
```
